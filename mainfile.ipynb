# 1. Names of most visited place:-

```python
options = [(f"User {i.name[:-4]}", i.path) for i in os.scandir("final/p1")]
dropdown1 = widgets.Dropdown(options=options, description='Select User:')

def update_output(user):
    dt = pd.read_csv(user)
    plt.figure(figsize=(8, 6))
    plt.bar(dt["name"][:15], dt["count"][:15], color='skyblue')
    plt.xlabel('Names')
    plt.ylabel('Counts')
    plt.title('Counts of Names')
    plt.xticks(rotation=60)
    plt.show()
interact(update_output, user=dropdown1)

```

# 3 places from where people come and go :-

```python
options = [(f"User {i.name[:-4]}", i.path) for i in os.scandir("final/p3")] 
dropdow = widgets.Dropdown(options=options, description='Select User:')
dropdown3 = widgets.Dropdown(description='Select User:')
dt51 = None
def update_output3(change):
    if change['type'] == 'change' and change['name'] == 'value':
        user = change['new']
        global dt51
        with open(user, 'rb') as f:
            dt51 = pk.load(f)
        options3 = [i for i in dt51.keys()]
        dropdown3.options = options3
        display(options3)
dropdow.observe(update_output3)




def update_output(u,place):
    key = []
    cnt = []
    if dt51 == None:
        return
    for k, v in sorted(dt51[place].items(), key=lambda item: item[1] , reverse=True):
        key.append(str(k))
        cnt.append(v)
    plt.figure(figsize=(8, 6))
    plt.bar(key[:15], cnt[:15], color='skyblue')
    plt.xlabel('Names')
    plt.ylabel('Counts')
    plt.title('Counts of Names')
    plt.xticks(rotation=60)
    plt.show()
interact(update_output, place=dropdown3,u=dropdow)
```

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
df = pd.read_csv("out/all.csv")
def print_path(n1,n2):
    cnt = {}
    if dt[n1].get(n2) == None:
        return
    for id ,l in enumerate(dt[n1].get(n2)):
        for i in range(id,len(dt[n1][n2])):
            if l[0] == dt[n1][n2][i][0]:
                cnt[id] = cnt.get(id,0) + 1
    k = max(cnt , key = cnt.get)
    ndf = df.loc[dt[n1][n2][k][1]:dt[n1][n2][k][2]]
    fig = px.scatter_mapbox(ndf, lat='lat', lon='long',  zoom=10)
    fig.add_trace(go.Scattermapbox(
        lat=ndf['lat'],
        lon=ndf['long'],
        mode='lines+markers'
    ))
    fig.update_layout(mapbox_style="open-street-map", mapbox_zoom=10, mapbox_center={"lat": ndf['lat'].mean(), "lon": ndf['long'].mean()},width = 1300,height = 600)
    fig.show()
```
